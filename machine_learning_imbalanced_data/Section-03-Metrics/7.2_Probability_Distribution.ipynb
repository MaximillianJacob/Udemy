```python
def load_balanced_data(num_obs, random_state):
    X, y = load_breast_cancer(return_X_y=True)
    X = pd.DataFrame(X)
    X['target'] = y
    
    X_bal = pd.concat([
        X[y==0].sample(num_obs, random_state = random_state),
        X[y==1].sample(
            num_obs, random_state=random_state)        
    ], axis=0)

    y = X_bal['target']
    y = y.map({0:1, 1:0})
    X = X_bal.drop('target', axis=1)

    return X.reset_index(drop=True), y.reset_index(drop=True)
```

```python
X, y = load_balanced_data(num_obs=100, random_state=10)

X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# separate dataset into train and set

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.3, random_state=0)

X_train.shape, X_test.shape
```

# Create imbalanced dataset

```python
def load_imbalanced_data():
    X, y = load_breast_cancer(return_X_y=True)
    X = pd.DataFrame(X)
    X['target'] = y

    X_bal = pd.concat([
        X[y==0].sample(100, random_state=0), 
        X[y==1].sample(10000, random_state=0, replace=True)
    ], axis=0)

    y = X_bal['target']
    y = y.map({0:1, 1:0})
    X = X_bal.drop('target', axis=1)

    return X.reset_index(drop=True), y.reset_index(drop=True)
```

```python
X_im , y_im = load_imbalanced_data()

X_im.head()
```

```python
#Verifying balancing ratio

y_im.value_counts(normalize=True)
```

```python
y_im.value_counts() /len(y_im)
```

```python
X_train_im, X_test_im, y_train_im, y_test_im = train_test_split(
                                                                X_im,
                                                                y_im,
                                                                test_size = 0.3,
                                                                random_state=0
)
```

# Train Logistic Regression on both datasets

```python
from sklearn.linear_model import LogisticRegression
```

```python
?LogisticRegression
```

```python
lr = LogisticRegression(
    penalty='l2',
    solver='newton-cg',
    C=1,
    random_state=0,
    max_iter=20).fit(X_train, y_train)
```

```python
probs = lr.predict_proba(X_test)[:,1]
```

```python
import sklearn.metrics
```

```python
dir(sklearn.metrics)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
help(roc_auc_score)
```

```python
#The correct approach, to use the probability value and not the predicted value
roc_auc_score(y_test, lr.predict_proba(X_test)[:,1])
```

```python
roc_auc_score(y_test, lr.predict(X_test))
```

```python
lr_im = LogisticRegression(
                    penalty='l2',
                    solver='newton-cg',
                    C=1,
                    random_state=0,
                    max_iter=20
)
lr_im.fit(X_im, y_im)
```

```python
probs_im = lr_im.predict_proba(X_test_im)[:,1]
```

```python
roc_auc_score(y_test_im, probs_im)
```

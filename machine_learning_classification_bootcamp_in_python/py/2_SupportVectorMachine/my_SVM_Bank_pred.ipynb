# Step 4: Model Training

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

# Unscaled

```python
df.head()
```

```python
X = df.drop(columns=['Customer ID', 'Retire'])
X.head()
```

```python
y =df.Retire
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.svm import SVC
```

```python
unscaled_svm = SVC()
unscaled_svm.fit(X_train, y_train)
```

```python
y_test_predict_unscaled = unscaled_svm.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test, y_test_predict_unscaled)
sns.heatmap(cm, annot=True, cmap='viridis')
```

```python
print(classification_report(y_test, y_test_predict_unscaled))
```

GP for optimization step:
1. Setup X and y variables
2. Perform feature scaling (verify data distribution remains the same)
3. Perform train_test_split on scaled data
4. Model training
5. Model prediction
6. confusion matrix, classification report

```python
X.head()
```

```python
min_X = X.min()
min_X
```

```python
max_range_X = (X - min_X).max()
max_range_X
```

```python
X_scaled = (X - min_X) / max_range_X
X_scaled
```

```python
plt.figure(figsize=(12,6))
plt.subplot(121)
sns.scatterplot(x='Age', y='401K Savings', data=X, hue=y, palette='viridis')

plt.subplot(122)
sns.scatterplot(x='Age', y='401K Savings', data=X_scaled, hue=y, palette='inferno')
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=1234)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.svm import SVC
```

```python
scaled_svm = SVC()
```

```python
scaled_svm.fit(X_train, y_train)
```

```python
y_test_predict_scaled = scaled_svm.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test, y_test_predict_scaled)
sns.heatmap(cm, annot=True, cmap='mako')
```

```python
print(classification_report(y_test, y_test_predict_scaled))
```

```python
plt.figure(figsize=(15,6))
plt.subplot(121)
sns.heatmap(confusion_matrix(y_test, y_test_predict_unscaled), annot=True, cmap='inferno',cbar=False)
plt.title('Unscaled Data')

plt.subplot(122)
sns.heatmap(cm, annot=True, cmap='viridis',cbar=False)
plt.title('Scaled Data')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../kdd2004.csv')
df.head()
```

```python
df.columns
```

```python
df.target.unique()
```

```python
df.target = df.target.map({-1:0, 1:1})
df.target.unique()
```

```python
help(df.target.value_counts)
```

```python
df.target.value_counts(normalize = True)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
X = df.drop('target', axis=1)
X.head()
```

```python
y = df.target
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0, test_size=0.3)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

# Training the model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
?RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 100, criterion='entropy', max_depth=2, random_state=0, n_jobs=-1)
rf.fit(X_train, y_train)
```

```python
y_train_rf = rf.predict_proba(X_train)[:,1]
y_train_rf[:5]
```

```python
y_test_rf = rf.predict_proba(X_test)[:,1]
y_test_rf[:5]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
?LogisticRegression
```

```python
lr = LogisticRegression(max_iter=1000, random_state=0)
lr.fit(X_train, y_train)
```

```python
y_train_lr = lr.predict_proba(X_train)[:,1]
y_test_lr = lr.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_test_base = pd.Series(np.zeros(len(y_test)))
y_test_base.head()
```

```python
print(f"""
Test Accuracy Baseline: {accuracy_score(y_test, y_test_base)}
Test Accuracy RF: {accuracy_score(y_test, rf.predict(X_test))}
Test Accuracy LR: {accuracy_score(y_test, lr.predict(X_test))}
      """)
```

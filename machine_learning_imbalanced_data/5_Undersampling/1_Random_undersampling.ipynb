```python
def make_data(sep):
    X, y = make_classification(n_samples=1000,
                               n_features=2,
                               n_redundant=0,
                               n_clusters_per_class=1,
                               weights=[0.66],
                               class_sep=sep,
                               random_state=0)
    
    #Transform arrays into pandas df and series
    X = pd.DataFrame(X, columns=['varA', 'varB'])
    y = pd.Series(y)

    return X, y
```

```python
#make datasets with different class separateness and plot

for sep in [0.1, 1., 2.]:
    X, y = make_data(sep)

    print(y.value_counts())

    sns.scatterplot(
        data=X, x='varA', y='varB', hue=y)
    
    plt.title('Separation: {}'.format(sep))
    plt.show()
```

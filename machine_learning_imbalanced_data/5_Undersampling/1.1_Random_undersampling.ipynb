```python
def make_data(sep):
    X, y = make_classification(n_samples=1000,
                           n_features=2,
                           n_redundant=0,
                           n_clusters_per_class=1,
                           weights=[0.99],
                           class_sep=sep,
                           random_state=1)
    
    #Transform arrays into pandas df and series
    X = pd.DataFrame(X, columns=['varA', 'varB'])
    y = pd.Series(y)

    return X, y
```

```python
#Separation cannot be 0.0
#This needs to be anything greater than 0

for sep in [0.01,1,2]:

    X, y = make_data(sep)

    print(y.value_counts())

    sns.scatterplot(
        data = X, x='varA', y='varB', hue=y
    )

    plt.title('Separation: {}'.format(sep))
    plt.show()
```

```python
from sklearn.metrics import roc_auc_score
```

```python
help(roc_auc_score)
```

```python
def run_randomForests(X_train, X_test, y_train, y_test):

    rf = RandomForestClassifier(n_estimators=200, random_state=39, max_depth=4)
    rf.fit(X_train, y_train)

    print('Train set')
    pred = rf.predict_proba(X_train)[:,1]
    print('Random Forests roc-auc: {}'.format(roc_auc_score(y_train, pred)))

    print('Test set')
    pred=rf.predict_proba(X_test)[:,1]
    print('Random Forests roc-auc: {}'.format(roc_auc_score(y_test, pred)))
```

```python
run_randomForests(X_train, X_test, y_train, y_test)
```

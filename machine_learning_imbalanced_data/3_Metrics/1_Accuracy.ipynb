# Accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
print('Accuracy Baseline test: ',  accuracy_score(y_test, y_test_base))
print('Accuracy Random Forest test: ', accuracy_score(y_test, rf.predict(X_test)))
print('Accuracy Logistic Regression test:', accuracy_score(y_test, lr.predict(X_test)))
```

# Percentage of the minority correctly classified

```python
def return_minority_perc(y_true, y_pred):
    minority_total = np.sum(y_true)
    condition = (y_true == 1) & (y_pred == 1)
    minority_correct =  np.sum(np.where(condition,1,0))
    return minority_correct/minority_total * 100
```

```python
print('% minority correct classified, Baseline test: ', return_minority_perc(y_test, y_test_base))
print('% minority correctly classifier, Random Forest test:', return_minority_perc(y_test, rf.predict(X_test)))

print('% minority correctly classifier, Regression test:', return_minority_perc(y_test, lr.predict(X_test)))
```
